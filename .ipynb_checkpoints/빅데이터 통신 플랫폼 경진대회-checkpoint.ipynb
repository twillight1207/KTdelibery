### 데이터 전처리

```python
# 배달 접수 시간 결측치 확인
dlvr['DLVR_RCEPT_TIME'].isnull().sum()
```

```python
# 배달 날짜와 배달 접수 시간 데이터 타입 변경
dlvr['PROCESS_DT'] = pd.to_datetime(dlvr['PROCESS_DT'])
dlvr['DLVR_RCEPT_TIME'] = pd.to_datetime(dlvr['DLVR_RCEPT_TIME'], format='%Y-%m-%d %H:%M:%S')
```

### 시간대 나누기
- 아침 6시~12시
- 점심 12시~18시
- 저녁 18시-24시
- 밤 24시~6시


```python
morning = dlvr[(dlvr['DLVR_RCEPT_TIME'].dt.hour >= 6)&(dlvr['DLVR_RCEPT_TIME'].dt.hour <= 11)].reset_index()
lunch = dlvr[(dlvr['DLVR_RCEPT_TIME'].dt.hour >= 12)&(dlvr['DLVR_RCEPT_TIME'].dt.hour <= 17)].reset_index()
dinner = dlvr[(dlvr['DLVR_RCEPT_TIME'].dt.hour >= 18)&(dlvr['DLVR_RCEPT_TIME'].dt.hour <= 23)].reset_index()
night = dlvr[(dlvr['DLVR_RCEPT_TIME'].dt.hour >= 0)&(dlvr['DLVR_RCEPT_TIME'].dt.hour <= 5)].reset_index()
```

```python
morning['Hours'] = 'morning'
lunch['Hours'] = 'lunch'
dinner['Hours'] = 'dinner'
night['Hours'] = 'night'
```

```python
morning.head()
```

```python
# 각 데이터프레임 합치키
dlvr_hours = pd.concat([morning, lunch, dinner, night])
```

## 시간대별 배달 주문량 변화 그래프

```python
dlvr_hours_count = pd.DataFrame(dlvr_hours.groupby(['Hours', 'PROCESS_DT'])['SN'].count().reset_index())
```

```python
dlvr_hours_count
```

```python
line,ax = plt.subplots(figsize=(15,6))
ax = sns.lineplot(x="PROCESS_DT", y="SN", data=dlvr_hours_count ,palette="pastel", hue="Hours")
ax.set_title("Delibery insight", fontsize=15)
ax.set_xlabel ("Date")
ax.set_ylabel ("Count")
ax.legend (loc="upper right")
```

- 3월의 첫 날에 주문건이 폭등함.
- 변화 주기는 시간대와 상관없이 유사한 모양을 가짐
- 저녁, 점심, 아침, 밤 순으로 그 수가 작아짐
- 저녁과 점심은 근소하게 증가, 아침 배달은 상당히 증가, 밤 배달은 거의 변화 없음

## 시간대별 배달 종류에 따른 주문량 변화

### 아침

```python
morning_count = pd.DataFrame(morning.groupby(['PROCESS_DT', 'Hours', 'DLVR_STORE_INDUTY_NM'])['SN'].count()).reset_index()
```

```python
plt.figure(figsize=(15,5))
plt.plot(morning_count['PROCESS_DT'], morning_count['SN'])
```

### 저녁
